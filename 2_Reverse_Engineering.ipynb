We are going to create a visual representation of the database, known as an ERD (Entity Relationship Diagram).  We are doing this early in the week because the ERD can help you understand how a database is constructed which makes querying it easier.<br></p><h2>What is an Entity Relationship Diagram?</h2>
<p>An Entity Relationship Diagram is a visual representation of the information in the database and how it is related.  It is a "snapshot" of the database architecture that can help you understand how the information you are working with is related.  An <u>entity</u> is essentially anything that can have information stored about it: It can be a person, thing, concept or event. In an ERD, each entity is has its own table with entity as the title. The information stored for that entity is called an <u>attribute</u>.  The attributes are included under the title in each table. Note that in a table of data, each attribute would be a column.</p>
<p>There are several variations of ERDs, but we are going to focus on the type we can create in MySQL Workbench.</p>

<p>Reverse engineering a database is the process of creating a visual model of an existing database.  In a future lesson, we will  discuss forward engineering which is the opposite process (creating a database based on a visual model).</p><p>The ease with which you can reverse (or forward) engineer a database is a notable advantage of using MySQL Workbench.</p><h2>Reverse Engineering in MySQL Workbench</h2><p>Click on the Database tab at the top of the screen and select "Reverse Engineer"<br></p>
<figure></figure>
<p>A window will pop up with a series of steps shown on the left. The instructions will tell you what to do at each step.</p>
<figure></figure>
<ul><li><strong>Connection Options</strong>: Keep default settings and hit "Next/Continue".</li><li><strong>Connect to DBMS:</strong>  You will be asked for your password. Enter it and hit "ok".  Then hit "Next/Continue".</li><li><strong>Select Schemas: </strong>Make sure the box with the "world" database is selected. Hit "Next/Continue".</li><li><strong>Retrieve Objects</strong>: Hit "Next/Continue".</li><li><strong>Select Objects:</strong> Hit "Execute".</li><li><strong>Reverse Engineer:</strong> It will say "Operation Completed Successfully".  Then hit "Next/Continue".</li><li><strong>Results:</strong> Hit "Finish".</li></ul>

<p><strong>Overview of the ERD</strong></p><p>The process you just completed created a visual model of the existing "world" database.  This visualization is known as an Entity Relationship Diagram (ERD). Notice that there is a table in the diagram for each table in the database (countries, languages, cities). Also note that the columns of each table are listed under the name of the table.  We will explore ERDs in more detail later in the course.</p>
<p>For now, it will help you to see an overview of how this database is designed and how the tables relate to each other. You can save this diagram by going to File &gt;Save Model AS. We suggest you save it in a designated folder.<br></p><p>You can always find your models by clicking on the model icon on the homepage.</p><figure></figure><p><br></p><p>You can also print the model to PDF by going to File &gt;Print to File.

<h2>SELECT *</h2>
<p>Using the SELECT keyword with the * operator chooses all of the columns in the table.</p>
<pre>SELECT * 
FROM countries;</pre>
<figure></figure>
<h2></h2>
<h2>LIMIT</h2>
<p>Anytime you want to set a limit on how many rows to return, you can use LIMIT.  To limit the output above to just the first 5 rows, use:</p>
<pre>SELECT * 
FROM countries
LIMIT 5;
</pre>
<figure></figure>
<p><br></p>
<h2>SELECT</h2>
<p>Rather than selecting all of the columns from a table, you can indicate which column(s) to select instead of using the * operator.  To select just the name and continent for all of the countries, use the following:</p>
<pre>SELECT name, continent
FROM countries;
</pre>
<figure></figure>
<p><br></p>
<h2>ORDER BY</h2>
<p>ORDER BY will sort strings alphabetically and numerical values from low to high.</p><p>To return the name  and continent of each country sorted alphabetically by name, add an ORDER BY clause.</p>
<pre>SELECT name, continent
FROM countries
ORDER BY name;
</pre>
<figure></figure>
<p><strong>DESC</strong></p>
<p>To put in opposite alphabetical order or descending numerical order add DESC to the ORDER BY statement.</p>
<pre>SELECT name, continent
FROM countries
ORDER BY name DESC;
</pre>
<figure></figure><p><br></p>
<h2>DISTINCT</h2>
<p>To produce a list of all of the continents without duplicates, we use SELECT DISTINCT.</p>
<pre>SELECT DISTINCT continent
FROM countries;
</pre>
<figure></figure>
<p><br></p>
<h2>WHERE</h2>
<p>Adding a WHERE clause allows you to filter the data and specify which rows you want to include.</p>
<p>To select only the countries in Asia, use the following:</p>
<pre>SELECT name, continent
FROM countries
WHERE continent = 'Asia';
</pre>
<figure></figure><p><br></p><h2>Advanced WHERE using AND/OR</h2>
<p>More advanced WHERE statements can be written that include AND/OR.  </p>
<p>For example, to list all the countries where the continent is Asia or Europe, write the following:</p>
<pre>SELECT name, continent
FROM countries
WHERE continent = 'Asia' OR continent = 'Europe';
</pre>
<figure></figure>

<h2>GROUP BY</h2>
<p>GROUP BY allows you to group rows based on a particular property so that an aggregate function can be applied to each group.  To find the maximum population for a country within each continent:</p>
<pre>SELECT continent, MAX(population)
FROM countries
GROUP BY continent;
</pre>
<figure></figure><p><br></p>
<h2>AS</h2>
<p>You have the option to provide a name (alias) to columns, expressions, or tables.  For example, if you wished to rename the output column for the MAX(population) to Max_Pop, use:</p>
<pre>SELECT name, continent, MAX(population) AS Max_Pop
FROM countries
GROUP BY continent;
</pre>
<figure></figure>
<h2>HAVING</h2>
<p>To add a filter to groups, apply a HAVING clause. This is similar to a WHERE clause, but you must use HAVING with when filtering groups.  To only include the most populated country from each continent where the country has at least a population of 100,000,000.</p>
<pre>SELECT continent, MAX(population) AS Max_Pop
FROM countries
GROUP BY continent
HAVING Max_Pop &gt; 100000000;
</pre>
